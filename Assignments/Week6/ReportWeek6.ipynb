##Abstraction
Cho $2$ chuỗi $A$ và $B$, tìm xem có bao nhiêu chuỗi con 2 phần tử liên tiếp của $A$ nằm $B$?

##Pattern Recognition
Tìm các chuỗi con của chuỗi này có nằm trong chuỗi kia hay không.

##Algorithm Design
Ý tưởng: 
- Tìm tất cả các chuỗi con của A và xem nó có thuộc B hay không?
- Để thực hiện bài toán nhanh nhất, ta lưu các chuỗi con vào 1 set để thực hiện tìm kiếm.
- Dùng cấu trúc dữ liệu $set$ của python để lưu các chuỗi con vì tìm kiếm trên $set$ có độ phức tạp $O(1)$.

Mã giả:
```
    input A, B
    initialize set_A contains all substring of A.
    initialize result = 0
    for substring in B:
        if substring in set_A: 
            result += 1
```

##Programming

````
A = input()
B = input()

set_b = {B[i] + B[i+1] for i in range(len(B) - 1)}

result = [A[i] + A[i+1] in set_b for i in range(len(A) - 1)]

print(sum(result))
````
