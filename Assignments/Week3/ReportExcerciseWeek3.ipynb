## **Đề bài: H-index**
Làm thế nào để đánh giá sự thành công của một nhà khoa học?  Dựa vào số bài báo được công bố hay dựa vào số lần một bài báo được trích dẫn tới ở công trình của những người khác? Cả hai tham số đó đều quan trọng.

Một bài báo có điểm số trích dẫn là c nếu nó được trích dẫn tới c lần trong các công trình của những nhà khoa học khác. Một trong số các cách đánh giá sự thành công của một nhà khoa học là tính chỉ số ảnh hưởng H_Index dựa trên sự kết hợp giữa số lượng bài báo và chỉ số trích dẫn của các bài báo đó.

Chỉ số H_Index của một nhà khoa học bằng k lớn nhất nếu người đó có k bài báo, mỗi bài có điểm số trích dẫn không nhỏ hơn k. Ví dụ, một người có 10 bài báo, mỗi bài báo được trích dẫn không dưới 10 lần thì H_Index của người đó ít nhất là bằng 10.
Một người có n bài báo, bài báo thứ i có điểm trích dẫn là $c_{i}$, i = 1 ÷ n. Hãy xác định H_Index của người đó.

Input
* Dòng đầu tiên chứa một số nguyên n ($1 ≤ n ≤ 5*10^5$),
* Dòng thứ 2 chứa n số nguyên $c_1, c_2, . . ., c_n (0 ≤ c_i ≤ 10^6, i = 1 ÷ n)$.

Output: đưa ra thiết bị xuất chuẩn một số nguyên – H_Index tìm được.

## Ví dụ

| INPUT                        |OUPUT              |
|------------------------------|-------------------|
|   5                          |     4             |
|   8 5 3 4 10                 |                   |


Giải thích : 
* Có 5 số nguyên. 
* Có 4 bài báo có số lần trích dẫn ít nhất là 4 và số bài báo còn lại là 3 < 4. 
* Vì vậy kết quả là 4



## ***Abstraction***
Cho dãy A có n số nguyên $c_1, c_2, . . ., c_n (0 ≤ c_i ≤ 10^6, i = 1 ÷ n)$. Tìm $k$ lớn nhất với điều kiện trong dãy A có $k$ phần tử ≥ $k$

## ***Algorithm Design***
### Ý tưởng
1.   Sort lại dãy input theo thứ tự tăng dần.
2.   Lặp qua mảng vừa sort từ theo thứ tự tăng dần, nếu số lượng phần tử còn lại của mảng bằng với phần tử đang xét thì in ra kết quả 

### Độ phức tạp O(N)

### Mã giả
```
input citations
sort citaions in ascending order
for i in len(citations):
    if len(citations) - i <= citations[i]:
        return len(citations) - i
```




## ***Programming***

````
citations = [8, 5, 3, 4, 10]
citations 
````

### Step 1: Sort lại dãy input theo thứ tự tăng dần.

````
citations.sort() 
citations
````

### Step 2: Lặp qua mảng vừa sort từ theo thứ tự tăng dần, nếu số lượng phần tử còn lại của mảng bằng với phần tử đang xét thì in ra kết quả.

````
for i, cited in enumerate(citations): 
        # tìm phần tử hiện tại
        result = len(citations) - i 
          
        # nếu result <= cited thì trả về kết quả
        if result <= cited: 
            print(result)
            break
````

````
n = int(input())
citations = input().split(' ')
citations = [int(x) for x in citations]

def H_index(citations): 
    
    # Sắp xếp theo thứ tự tăng dần
    citations.sort() 
      
    # Lặp qua dãy số
    for i, cited in enumerate(citations): 
        # tìm phần tử hiện tại
        result = len(citations) - i 
          
        # nếu result <= cited thì trả về kết quả
        if result <= cited: 
            return result 
           
    return 0

print(H_index(citations))
````

## ***Algorithm Design***
### 1. Ý tưởng
#### a. Tìm các cụm liên thông
Áp dụng thuật toán DFS. DFS sẽ xây dựng 1 cây, trong đó xem từng node trong dồ thị là node gốc (nếu node đó chưa được xét qua). Thuật toán này sẽ mở rộng cây sâu nhất có thể dựa trên liên kết giữa các node.
<!-- DFS có độ phức tạp O(V + E), với V là số node và E là số cạnh. -->

#### b. Tìm số cách thực hiện liên kết giữa các cụm
Áp dụng công thức sau:

$S =$ $n^{k – 2} \times \prod_{i=1}^{k} s_i$(*)

Trong đó:

$n$: tổng số node trong đồ thị

$k$: Số cụm liên thông

$s_i$: số node trong cụm thứ i

Công thức (*) được tham khảo từ [đây](https://laptrinhx.com/srm-733-cayley-s-formula-derived-2556835044/)

## ***Programming***

````
n, m = map(int, input().split())

graph = {}
for _ in range(m):
    u, v = map(int, input().split())
    if u not in graph.keys():
        graph[u] = {v}
    else:
        graph[u].add(v)

    if v not in graph.keys():
        graph[v] = {u}
    else:
        graph[v].add(u)

def get_all_connected_groups(graph):
    already_seen = set()
    result = []
    for i in range(1, n + 1):
        if i not in already_seen:
            connected_group, already_seen = get_connected_group(i, already_seen)
            result.append(connected_group)
    return result


def get_connected_group(node, already_seen):
    index = 0
    queue = [node]
    already_seen.add(node)

    while index < len(queue):
        u = queue[index]
        index += 1
        if u not in graph:
            continue
        for v in graph[u]:
            if v not in already_seen:
                already_seen.add(v)
                queue.append(v)
    return queue, already_seen

com = get_all_connected_groups(graph)
print(len(com) - 1)

anw = 0
if len(com) > 1:
    base = int(1e9 + 7)
    anw = 1
    anw2 = 1
    for _ in range(len(com) - 2):
        anw2 = (anw2*n) % base
    for x in com:
        anw = (anw * len(x)) % base
    anw = anw * anw2 % base

print(anw)
````
