## **Đề bài: Đổi mới**
Thành phố quyết định chặt hạ hết n cây xanh hiện có trong thành phố để trồng chỉ một loại cây mới. Nhiệm vụ được giao cho Công ty Cây xanh thành phố. Do hạn chế về thiết bị Công ty chỉ tổ chức được 2 đội đốn hạ cây. Đội I hạ được $a$ cây mỗi ngày, nhưng cứ mối ngày thứ $k$ thì phải nghỉ để bảo dưỡng kỹ thuật, tức là đội I sẽ nghỉ vào các ngày $k$, 2$k$, 3$k$, . . . Đội II hạ được $b$ cây mỗi ngày, nhưng cứ mỗi ngày thứ $m$ thì phải nghỉ để bảo dưỡng kỹ thuật, tức là đội II sẽ nghỉ vào các ngày $m$, 2$m$, 3$m$, . . . Ở ngày nghỉ, số cây chặt hạ của đội sẽ là 0. Cả hai đội bắt đầu công việc vào cùng một ngày và làm việc song song với nhau.

Công việc trồng cây mới sẽ bắt đầu sau khi toàn bộ cây cũ đã bị đốn hạ.

Hãy xác định sau bao nhiêu ngày thì có thể bắt đầu việc trồng mới cây.

**Input:** 5 số nguyên $a$, $k$, $b$, $m$ và $n$ ($1 ≤ $a$, $b$ ≤ 10^9, 2 ≤ $k$, $m$ ≤ 10^{18}, 1 ≤ $n$ ≤ 10^{18}$)

**Output:** một số nguyên – số ngày tính được.

## **1. Abstraction**

Cho các số nguyên $a,k,b,m,n$ . Tìm số nguyên $x$ nhỏ nhất sao cho:
$$M(x) = a \times (x - \lfloor \frac{x}{k} \rfloor) + b \times (x - \lfloor \frac{x}{m} \rfloor) \geqslant n $$
Trong đó:
* $x$ là số ngày chặt cây của cần tìm. Trong ngày thứ $x$, một trong hai đội có thể không nhất thiết làm đủ số cây mà đội đó có thể chặt trong một ngày.
* Vì đội I nghỉ vào các ngày $k$, $2k$, $3k$, ... nên cứ $x$ ngày thì đội I nghỉ $\lfloor \frac{x}{k} \rfloor$. ngày. Do đó trong $x$ ngày thì số ngày đội $I$ chặt được cây còn là $x - \lfloor \frac{x}{k} \rfloor$. Tương tự đội II làm việc trong $x - \lfloor \frac{x}{m} \rfloor$ ngày.
* $M(x)$ là tổng số cây lớn nhất mà cả hai đội tối đa làm được cả $x$ ngày ( làm đủ số cây trong ngày thứ $x$ ). Công việc chặt cây kết thúc khi số cây chặt được $M(x)$ bắt đầu lớn hơn hoặc bằng số cây cần chặt $n$.

## **2. Pattern-Recognition**
Phương pháp 1: 
* Sử dụng thuật toán **linear search**. Ta sẽ duyệt từ $1,2,3,...$ cho đến khi nào $M(x) \geqslant n$, $x$ là kết quả cuối cùng.
* Độ phức tạp: $O(n)$

Phương pháp 2:
* Vì $M(x)$ luôn tăng khi $x$ tăng, ta có thể áp dụng thuật toán **Binary Search (tìm kiếm nhị phân)** để tìm $x$ nhỏ nhất thỏa điều kiện.
* Độ phức tạp: $O(log(n))$
* Tài liệu tham khảo: [Binary Search](https://www.geeksforgeeks.org/binary-search/)


## **3. Algorithm Design**
### **Ý tưởng**
Do $M(x)$ là luôn tăng khi $x$ tăng nên thỏa điều kiện để cài đặt thuật toán **Binary Search**:
* Khởi tạo: left = $0$ và right = $10^{18}$ ( $x < n \leqslant 10^{18}$ ).
* Đệ quy với điều kiện  $right \geqslant left$ , và xét $ mid = left + \lfloor \frac{right-left}{2} \rfloor$.
* Nếu $M$(mid) $ > n$, thì $mid$ là chặn trên nhỏ nhất của $x$ lúc này, vì thế $x$ cần tìm nằm trong khoảng [left,mid - 1] và ta lưu lại giá trị chặn trên này cho lần chạy sau (nếu thuật toán dừng lại, giá trị này chính là kết quả của bài toán), ngược lại nếu $M$(mid) $ < n$ thì $x$ cần tìm nằm trong khoảng [mid+1, left]. Tiếp tục chạy lại đệ quy với miền giá trị mới cho đến khi thuật toán kết thúc.
* Thuật toán dừng lại khi $M$(mid) = n hoặc $right < left$.

## **4. Cài đặt**





    

```
def binarySearch(xLeft,xRight,neareast):
  if xLeft <= xRight:
    xMid = xLeft + (xRight-xLeft) // 2
    nXMid = a*(xMid-xMid//k) + b*(xMid-xMid//m)
    if (nXMid == n):
      return xMid
    if (nXMid > n):
      return binarySearch(xLeft, xMid - 1, xMid)
    else:
      return binarySearch(xMid + 1, xRight, neareast)
  else:
    return neareast


a, k, b, m, n = map(int, input().split())
result = binarySearch(0, 10**18, 10**18)
print(result)
```
