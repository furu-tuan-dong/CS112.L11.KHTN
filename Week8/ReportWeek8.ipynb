Đề bài: XÂU CON TƯƠNG ĐƯƠNG (1s, 50M)

Cho xâu s chỉ chứa các ký tự la tinh thường độ dài n không quá 105. Ký hiệu s[i..j] là xâu con các ký tự liên tiếp nhau của s từ vị trí i đến vị trí j. Hai xâu con x = s[a..b] và y = s[c..d] được gọi là cùng một lớp tương đương nếu có cách đổi chỗ các ký tự trong một xâu để nhận được xâu kia. Các vị trí được đánh số từ 1 trở đi.

Ví dụ với s = “abcbacaac”, s[1..3] tương đương với s[4..6], còn s[1..3] không tương đương với s[5..7].

Cho q truy vấn, mỗi truy vấn là 4 số nguyên a, b, c, d xác định các xâu con x và y. Với mỗi truy vấn hãy xác định hai xâu con này có cùng một lớp tương đương hay không và đưa ra câu trả lời tương ứng “YES” hoặc “NO”.

Dữ liệu: Vào từ thiết bị nhập chuẩn:

Dòng đầu tiên chứa xâu s,

Dòng thứ 2 chứa số nguyên q (1 ≤  q  ≤ 105),

Mỗi dòng trong q dòng tiếp theo chứa 4 số nguyên a, b, c, d (1 ≤ a ≤ b ≤ n, 1 ≤ c ≤ d ≤ n).

Kết quả: Đưa ra thiết bị xuất chuẩn các câu trả lời xác định được cho mỗi truy vấn, mỗi câu trả lời trên một dòng.

Dữ liệu mẫu:

|INPUT|OUTPUT|
| :------------:|:-------------:|
|abcbacaac|YES|
|2|NO|
|1346||
|1357||


ABSTRACTION

Cho xâu $s$ và $2$ xâu con $a$, $b$. Kiểm tra $a$ và $b$ có tương đương nhau không?

PATTERN RECOGNITION

Kiểm tra 2 xâu tương đương.

ALGORITHM DESIGN

Ý tưởng: 

- Hai xâu a và b tương đương nhau nếu tổng số kí tự bất kì trong a bằng với tổng số kí tự đó trong b.
- Thay vì đếm số lần xuất hiện của từng kí tự, ta gán từng kí tự với một số nào đó, 2 xâu tương đương nhau nếu có tổng các số này bằng nhau.
- Để tránh trùng lẫn, kí tự c sẽ được gán số theo phép tính $1<<ord(c)$ trong python.
- Hàm $ord(c)$ trả về một số nguyên đại diện cho kí tự $c$ trong bảng ASCII. Phép tính $<<$: dịch bit sang trái.

Mã giả:

```
Input s, q.
Insert 0 at head of s.
For i from 1 to length s:
    Assign s[i] = s[i-1] + 1 << ord(character)
For i from 0 to q:
    Input a, b, c ,d.
    if s[b] - s[a-1] = s[d] - s[c-1]:
        print YES
    else:
        print NO
```

PROGRAMMING

````
from sys import stdin,stdout
s = input()
s = [0] + list(s.strip())
q = int(input())

for i in range(1, len(s)):
    s[i] =  s[i-1] + (1<<ord(s[i]))

for _ in range(q):
    a,b,c,d = map(int, input().split())
    if s[b] - s[a-1] == s[d] - s[c-1]:
        print('YES')
    else:
        print('NO')
````
